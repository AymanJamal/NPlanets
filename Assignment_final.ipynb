In this following code, I will be checking consistency in the data and to check if there is any missing values

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)

# Identify columns with missing values
columns_with_missing_values = missing_values[missing_values > 0].index
print("Columns with missing values:")
print(columns_with_missing_values)

# Print the shape of the dataset
print("Shape of the dataset:")
print(df.shape)
```

The below code can be used to repalce the missing values with the mean of each culomn. We repalce the missing values with the mean so that the data would not get affected.

```python
# Load the data
df_replacemissing = pd.read_csv("Scotland_teaching_file_1PCT.csv")

# replace missing values with mean for each numeric column
for col in df_replacemissing.select_dtypes(include=['float64', 'int64','string']):
    df_replacemissing[col].fillna(df_replacemissing[col].mean(), inplace=True)

# Print the dataset after replacing missing values
print("\nDataset after replacing missing values with their mean:")
df_replacemissing.head()
```

The below code can be used to remove any duplicated rows in the dataset

```python
# Removing duplicate rows
df_replacemissing = df_replacemissing.drop_duplicates()

# Verifying if duplicates are removed
remaining_duplicates = df_replacemissing.duplicated().sum()

# Display the result
remaining_duplicates, df_replacemissing.shape
print(remaining_duplicates, df_replacemissing.shape)
```

Upon running the code it shows as there was 0 duplicated record in this dataset.
